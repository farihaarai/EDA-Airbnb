**#identifying outliers in price**

```python
# df = data[data['price'] < 1500]
# sns.boxplot(data = df,x = 'price')

# Step 1: Filter data for listings with prices below 1500
df = data[data['price'] < 2000]

# Step 2: Calculate the boundaries for detecting outliers
Q1 = df['price'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df['price'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile Range

# Define outliers as points outside 1.5 times the IQR
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Step 3: Identify rows where 'price' is an outlier
outliers = df[(df['price'] < lower_limit) | (df['price'] > upper_limit)]

# Step 4: Plot the boxplot
sns.boxplot(data=df, x='price')

# Step 5: Print the number of outliers
print(f"Number of outliers: {len(outliers)}")
```

```python
import matplotlib.pyplot as plt
df['room_type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Range by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('neighbourhood')['availability_365'].mean().nlargest(5).plot(kind='bar', color='blue')
plt.title('Top 5 Neighbourhoods by Availability')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Availability (Days)')
plt.xticks(rotation=45)
plt.show()
```

```python
df['host_name'].value_counts().nlargest(10).plot(kind='bar', color='pink')
plt.title('Top 10 Hosts by Number of Listings')
plt.xlabel('Host Name')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Price Variation by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
reviews_over_time = df.groupby(df['last_review'].dt.to_period('M'))['number_of_reviews'].sum()
reviews_over_time.plot(kind='line', color='red')
plt.title('Trend of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
bins = [0, 50, 100, 200, 500, 1000]
labels = ['<50', '50-100', '100-200', '200-500', '500+']
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels)
df.groupby('price_range')['availability_365'].mean().plot(kind='bar', color='cyan')
plt.title('Average Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Availability')
plt.show()
```

```python
df.groupby('room_type')['number_of_reviews'].mean().plot(kind='bar', color='orange')
plt.title('Average Number of Reviews by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Number of Reviews')
plt.xticks(rotation=45)
plt.show()
```
